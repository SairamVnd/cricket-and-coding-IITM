```python
#creating a dataframe 
team=labels.values
df_combined=pd.DataFrame()
for i in team:
    print("------------",i,"-----------")
    for j in team:
        if i!=j:
            print(j)
            df_iteration=df_bat_field_team.loc[(df_bat_field_team["Team1"] ==i) & (df_bat_field_team["Team2"] == j) | (df_bat_field_team['Team1'] ==j) & (df_bat_field_team['Team2'] == i)]
            df_combined=pd.concat([df_combined,df_iteration])

            

```

```python
df_combined
```

```python
# Iterate over each row in the dataframe
for i, row in df_combined.iterrows():
    # Check if the toss decision is "bat"
    if row['TossDecision'] == 'bat':
        # Assign the toss winner as the batting team and the other team as the fielding team
        df_combined.at[i, 'Batting Team'] = row['TossWinner']
        df_combined.at[i, 'Fielding Team'] = row['Team2'] if row['Team1'] == row['TossWinner'] else row['Team1']
    # If the toss decision is "field"
    elif row['TossDecision'] == 'field':
        # Assign the toss winner as the fielding team and the other team as the batting team
        df_combined.at[i, 'Fielding Team'] = row['TossWinner']
        df_combined.at[i, 'Batting Team'] = row['Team2'] if row['Team1'] == row['TossWinner'] else row['Team1']

# Drop the columns "TossWinner" and "TossDecision"
#df_combined = df_combined.drop(['TossWinner', 'TossDecision'], axis=1)
df_combined
```

```python

df_b2b_score_6overs = pd.DataFrame({
    "ID": df_b2b_6overs["ID"],
    "innings": df_b2b_6overs["innings"],
    "overs": df_b2b_6overs["overs"],
    "total_run": df_b2b_6overs["total_run"]
})

df_b2b_score_6overs
```

```python

# Filter the dataset to only include innings 1
innings1_df = df_b2b_score_6overs[df_b2b_score_6overs["innings"] == 1]

# Group the dataset by match ID and calculate the sum of total runs
sum_score_df = innings1_df.groupby("ID")["total_run"].sum().reset_index()

# Create a new data frame with columns ID, innings, and sum score
new_df1 = pd.DataFrame({"ID": sum_score_df["ID"], "innings": 1, "sum_score": sum_score_df["total_run"]})

# Display the new data frame
(new_df1)
```

```python
# Filter the dataset to only include innings 2
innings2_df = df_b2b_score_6overs[df_b2b_score_6overs["innings"] == 2]

# Group the dataset by match ID and calculate the sum of total runs
sum_score_df = innings2_df.groupby("ID")["total_run"].sum().reset_index()

# Create a new data frame with columns ID, innings, and sum score
new_df2 = pd.DataFrame({"ID": sum_score_df["ID"], "innings": 2, "sum_score": sum_score_df["total_run"]})

# Display the new data frame
(new_df2)
```

```python
# filter the DataFrame based on fielding and batting teams

batting_team=input("enter name of batting team: ")
fielding_team=input("enter name of fielding team: ")
filtered_df1 = df[(df['Fielding Team'] == fielding_team ) & (df['Batting Team'] == batting_team)]

# plot the data

plt.figure(figsize=(10, 5))
plt.plot(filtered_df1['Season'], filtered_df1['sum_score'])
plt.xlabel('Season')
plt.ylabel('Total Score')
plt.title(f'bat={batting_team} vs field={fielding_team}')
plt.show()


#Data

x=filtered_df1.iloc[:,1:2].values
y=filtered_df1.iloc[:,5:]

#train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print ('x Shape-',x.shape)
print ('y Shape-', y.shape)
print ('x_Train Shape-',x_train.shape)
print ('Y_Train Shape-',y_train.shape)
print ('X_Test Shape-',x_test.shape)
print ('Y_Test Shape-',y_test.shape)

#model creation

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr=lr.fit(x_train,y_train)
lr

#prediction

y_pred=lr.predict(x_test)


df2=pd.DataFrame(np.hstack((y_pred,y_test,y_pred-y_test)),columns=["prediction","actual","difference"])
display(df2)

#Visualisation

plt.scatter(x_test,y_test )
plt.plot(x_test,y_pred,"r")

#r2_score
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,y_pred)
print(accuracy)


"""
TEAM NAMES

Chennai Super Kings
Delhi Capitals
Gujarat Titans
Kolkata Knight Riders
Lucknow Super Giants
Mumbai Indians
Punjab Kings
Rajasthan Royals
Royal Challengers Bangalore
Sunrisers Hyderabad"""
```
